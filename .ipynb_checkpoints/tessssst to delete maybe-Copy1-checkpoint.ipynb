````python
import numpy as np

def load_profits_from_file(filename, num_objectives, num_items):
    profits = np.zeros((num_objectives, num_items))
    with open(filename, 'r') as f:
        lines = [line.strip() for line in f if line.strip()]
    idx = 0
    idx += 1  # skip header
    for obj in range(num_objectives):
        idx += 1  # skip capacity line
        for item in range(num_items):
            idx += 1  # skip "n:" line
            idx += 1  # skip weight
            profit = int(lines[idx])
            profits[obj, item] = profit
            idx += 1
    return profits

def load_weights_from_file(filename, num_objectives, num_items):
    weights = np.zeros((num_objectives, num_items), dtype=int)
    with open(filename, 'r') as f:
        lines = [line.strip() for line in f if line.strip()]
    idx = 0
    idx += 1  # skip header
    for obj in range(num_objectives):
        idx += 1  # skip capacity line
        for item in range(num_items):
            idx += 1  # skip "n:" line
            weight = int(lines[idx])
            weights[obj, item] = weight
            idx += 2  # skip profit
    return weights

def load_capacities_from_file(filename, num_objectives, num_items):
    with open(filename, 'r') as f:
        lines = [line.strip() for line in f if line.strip()]
    idx = 0
    idx += 1  # skip header
    capacities = []
    for obj in range(num_objectives):
        cap = float(lines[idx])
        capacities.append(cap)
        idx += 1 + 3 * num_items
    return np.array(capacities)
````

````python
%%cython
import numpy as np
import random
import textwrap
import re
import ast
import difflib
from scipy.stats import entropy
from sklearn.metrics import mutual_info_score
from scipy.spatial import distance

def mutation_move(observation, profits, weights, capacities, archive_objs=None, best_observation=None):
    items_selected = np.where(observation == 1)[0]
    items_unselected = np.where(observation == 0)[0]
    if len(items_selected) == 0 or len(items_unselected) == 0:
        return [0], [0]
    
    # Multi-item support with adaptive selection
    nchange = min(3, len(items_selected), len(items_unselected), random.choice([1,2,3]))
    remove_idx = list(np.random.choice(items_selected, nchange, replace=False))
    add_idx = list(np.random.choice(items_unselected, nchange, replace=False))
    return remove_idx, add_idx

def local_search_move(observation, profits, weights, capacities, archive_objs=None, best_observation=None):
    items_selected = np.where(observation == 1)[0]
    items_unselected = np.where(observation == 0)[0]
    if len(items_selected) == 0 or len(items_unselected) == 0 or profits is None:
        return [0], [0]
    
    # Enhanced with weighted profit consideration
    profs_selected = np.sum(profits[:, items_selected], axis=0)
    profs_unselected = np.sum(profits[:, items_unselected], axis=0)
    
    nchange = min(3, len(items_selected), len(items_unselected), random.choice([1,2,3]))
    remove_idx = [items_selected[i] for i in np.argsort(profs_selected)[:nchange]]
    add_idx = [items_unselected[i] for i in np.argsort(profs_unselected)[-nchange:]]
    return remove_idx, add_idx

def global_search_move(observation, profits, weights, capacities, archive_objs=None, best_observation=None):
    items_selected = np.where(observation == 1)[0]
    items_unselected = np.where(observation == 0)[0]
    if len(items_selected) == 0 or len(items_unselected) == 0 or profits is None:
        return [0], [0]
    
    # Enhanced with probabilistic selection based on profit ratios
    profs_selected = np.sum(profits[:, items_selected], axis=0) + 1e-9
    profs_unselected = np.sum(profits[:, items_unselected], axis=0) + 1e-9
    
    nchange = min(3, len(items_selected), len(items_unselected), random.choice([1,2,3]))
    remove_idx = list(np.random.choice(items_selected, nchange, p=profs_selected/np.sum(profs_selected), replace=False))
    add_idx = list(np.random.choice(items_unselected, nchange, p=profs_unselected/np.sum(profs_unselected), replace=False))
    return remove_idx, add_idx

def follow_move(observation, profits, weights, capacities, archive_objs=None, best_observation=None):
    if best_observation is None:
        return mutation_move(observation, profits, weights, capacities)
    
    # Enhanced to follow elite solutions
    diff = best_observation - observation
    add_candidates = np.where((diff == 1) & (observation == 0))[0]
    remove_candidates = np.where((diff == -1) & (observation == 1))[0]
    
    if len(add_candidates) == 0 or len(remove_candidates) == 0:
        return mutation_move(observation, profits, weights, capacities)
    
    nchange = min(3, len(add_candidates), len(remove_candidates), random.choice([1,2,3]))
    add_idx = list(np.random.choice(add_candidates, nchange, replace=False))
    remove_idx = list(np.random.choice(remove_candidates, nchange, replace=False))
    return remove_idx, add_idx

def diversity_move(observation, profits, weights, capacities, archive_objs=None, best_observation=None):
    best_move = None
    best_min_dist = -np.inf
    items_selected = np.where(observation == 1)[0]
    items_unselected = np.where(observation == 0)[0]
    
    if len(items_selected) == 0 or len(items_unselected) == 0 or archive_objs is None or len(archive_objs) == 0:
        return [0], [0]
    
    # Enhanced with multiple candidate evaluation
    for _ in range(5):
        nchange = min(3, len(items_selected), len(items_unselected), random.choice([1,2,3]))
        remove_idx = list(np.random.choice(items_selected, nchange, replace=False))
        add_idx = list(np.random.choice(items_unselected, nchange, replace=False))
        
        candidate = observation.copy()
        candidate[remove_idx] = 0
        candidate[add_idx] = 1
        candidate_objs = np.sum(profits * candidate, axis=1)
        
        # Calculate diversity using Euclidean distance
        dists = np.linalg.norm(archive_objs - candidate_objs, axis=1)
        min_dist = np.min(dists)
        
        if min_dist > best_min_dist:
            best_min_dist = min_dist
            best_move = (remove_idx, add_idx)
    
    return best_move if best_move is not None else ([0], [0])

def novelty_check(existing_funcs, new_func_code, similarity_threshold=0.8):
    """Enhanced novelty check with AST comparison and semantic similarity"""
    try:
        new_ast = ast.parse(new_func_code)
        new_code_str = ast.unparse(new_ast)
        
        for old_code in existing_funcs:
            try:
                old_ast = ast.parse(old_code)
                old_code_str = ast.unparse(old_ast)
                
                # Calculate structural similarity
                structural_sim = difflib.SequenceMatcher(None, old_code_str, new_code_str).ratio()
                
                # If too similar structurally, reject
                if structural_sim > similarity_threshold:
                    return False
                    
            except Exception:
                continue
                
        return True
    except Exception:
        return False

MOVE_REGISTRY = {
    "mutation": mutation_move,
    "local_search": local_search_move,
    "global_search": global_search_move,
    "follow": follow_move,
    "diversity": diversity_move,
}
````

````python
%%cython
import numpy as np
import ast
import random
import textwrap
import requests

class ModularMetaAgent:
    def __init__(self, num_items, num_objectives, profits, weights, capacities, move_types=None, smoothing=0.1):
        self.num_items = num_items
        self.num_objectives = num_objectives
        self.profits = profits
        self.weights = weights
        self.capacities = capacities
        self.architecture = AgentArchitecture(move_types if move_types else list(MOVE_REGISTRY.keys()), smoothing=smoothing)
        self.llm_funcs = {}
        self.llm_stats = {}
        self.llm_code_history = []
        self.best_observation = None
        self.last_move_type = None
        self.last_fitness = []
        self.move_history = []  # Track sequence of moves
        self.outcome_history = []  # Track outcomes of moves
        
    def observe(self, state):
        return np.array(state["Items"], dtype=np.int32)
    
    def act(self, observation, context=None):
        archive_objs = context.get("archive_objs") if context else None
        best_observation = context.get("best_observation") if context and "best_observation" in context else self.best_observation
        
        # Calculate current solution diversity
        current_objs = np.sum(self.profits * observation, axis=1)
        diversity_score = self.calculate_diversity_score(current_objs, archive_objs)
        
        # Enhanced decision making with multiple strategies
        use_llm_move = False
        
        # Strategy 1: Use LLM moves when diversity is low
        if diversity_score < 0.5 and len(self.llm_funcs) > 0:
            use_llm_move = True
        
        # Strategy 2: Use LLM moves when stuck in local optima
        if len(self.last_fitness) > 5:
            if np.std(self.last_fitness[-5:]) < 0.01:
                use_llm_move = True
        
        # Strategy 3: Use LLM moves periodically for exploration
        if random.random() < 0.1 and len(self.llm_funcs) > 0:
            use_llm_move = True
        
        # Try to find a valid move
        for _ in range(3):
            if use_llm_move and len(self.llm_funcs) > 0:
                # Prioritize LLM moves
                llm_moves = [mt for mt in self.architecture.move_types if mt.startswith('llm_move_')]
                if llm_moves:
                    move_type = random.choice(llm_moves)
                else:
                    move_type = self.architecture.select_move(context)
            else:
                move_type = self.architecture.select_move(context)
            
            if move_type in MOVE_REGISTRY:
                move_fn = MOVE_REGISTRY[move_type]
                removes, adds = move_fn(observation, self.profits, self.weights, self.capacities, 
                                       archive_objs=archive_objs, best_observation=best_observation)
            elif move_type in self.llm_funcs:
                try:
                    removes, adds = self.llm_funcs[move_type](observation, self.profits, self.weights, 
                                                              self.capacities, archive_objs, best_observation)
                    self.llm_stats[move_type]["calls"] += 1
                except Exception as e:
                    print(f"LLM move error: {e}")
                    continue
            else:
                continue
            
            if self._is_feasible(observation, removes, adds):
                self.last_move_type = move_type
                self.move_history.append(move_type)
                return removes, adds
        
        # Fallback to mutation
        removes, adds = mutation_move(observation, self.profits, self.weights, self.capacities)
        self.last_move_type = "mutation"
        self.move_history.append("mutation")
        return removes, adds
    
    def report_move_result(self, move_type, success, fitness_delta=0):
        if move_type is None:
            move_type = self.last_move_type
        self.architecture.report_move_result(move_type, success, fitness_delta)
        self.outcome_history.append(success)
        self.last_fitness.append(fitness_delta)
        
        if move_type in self.llm_stats:
            if success:
                self.llm_stats[move_type]["successes"] += 1
    
    def update_best(self, observation):
        self.best_observation = observation.copy()
    
    def add_llm_move(self, code: str):
        """Enhanced LLM move integration with validation"""
        header = (
            "def llm_move(observation, profits, weights, capacities, archive_objs, best_observation=None):\n"
            '    """\n'
            "    Propose a multi-item move for multi-objective knapsack.\n"
            "    Returns: (remove_indices: list[int], add_indices: list[int])\n"
            '    """'
        )
        
        # Clean and format the code
        code = code.strip()
        if not code.startswith("def llm_move"):
            idx = code.find("def llm_move")
            if idx >= 0:
                code = code[idx:]
        
        # Extract only the function
        code_lines = code.splitlines()
        func_lines = []
        inside = False
        for line in code_lines:
            if line.strip().startswith("def llm_move"):
                inside = True
            if inside:
                func_lines.append(line)
            if inside and line.strip().startswith("def ") and len(func_lines) > 1:
                break
        code = "\n".join(func_lines)
        
        # Add strict docstring if missing
        if '"""' not in code:
            colon_idx = code.find(":")
            if colon_idx != -1:
                body = textwrap.dedent(code[colon_idx+1:])
                code = header + "\n" + textwrap.indent(body, "    ")
            else:
                code = header + "\n    pass"
        
        # Validate using improved novelty check
        if not novelty_check(self.llm_code_history, code):
            print("[DEBUG] LLM move rejected: not novel.")
            return None
        
        func = self._validate_and_compile_llm_func(code)
        if func is not None:
            move_name = f"llm_move_{len(self.llm_funcs)}"
            self.architecture.move_types.append(move_name)
            self.architecture.move_probs = np.append(self.architecture.move_probs, 0.1)
            self.architecture.move_probs /= np.sum(self.architecture.move_probs)
            self.llm_funcs[move_name] = func
            self.llm_stats[move_name] = {"calls": 0, "successes": 0}
            self.llm_code_history.append(code)
            print(f"[DEBUG] Added LLM move: {move_name}")
            return move_name
        print("[DEBUG] LLM move failed integration.")
        return None
    
    def prune_llm_moves(self, min_success_rate=0.05, min_calls=10):
        """Remove underperforming LLM moves"""
        to_remove = []
        for move_name, stats in self.llm_stats.items():
            calls = stats["calls"]
            successes = stats["successes"]
            rate = successes / calls if calls > 0 else 0
            if calls > min_calls and rate < min_success_rate:
                to_remove.append(move_name)
        
        for move_name in to_remove:
            if move_name in self.architecture.move_types:
                idx = self.architecture.move_types.index(move_name)
                self.architecture.move_types.pop(idx)
                self.architecture.move_probs = np.delete(self.architecture.move_probs, idx)
                self.architecture.move_probs /= np.sum(self.architecture.move_probs)
            self.llm_funcs.pop(move_name, None)
            self.llm_stats.pop(move_name, None)
    
    def _validate_and_compile_llm_func(self, code):
        """Enhanced validation with feasibility testing"""
        loc = {}
        
        # Add required imports if missing
        lines = code.splitlines()
        imports_needed = []
        code_lower = code.lower()
        if "entropy" in code_lower and "import entropy" not in code_lower:
            imports_needed.append("from scipy.stats import entropy")
        if "mutual_info_score" in code_lower and "import mutual_info_score" not in code_lower:
            imports_needed.append("from sklearn.metrics import mutual_info_score")
        if "distance" in code_lower and "import distance" not in code_lower:
            imports_needed.append("from scipy.spatial import distance")
        if imports_needed:
            code = "\n".join(imports_needed) + "\n" + code
        
        # Check for forbidden patterns
        entropy_calls = re.findall(r'entropy\s*\((.*?)\)', code)
        for call in entropy_calls:
            arg_count = len([x for x in call.split(',') if x.strip() != ''])
            if arg_count > 2:
                print("[DEBUG] LLM func rejected: entropy() called with >2 arguments.")
                return None
        
        # Syntax validation
        try:
            tree = ast.parse(code)
        except Exception as e:
            print("[DEBUG] LLM func rejected: invalid syntax", e)
            return None
        
        try:
            exec(code, {"np": np}, loc)
            llm_func = [v for v in loc.values() if callable(v)][0]
            
            # Enhanced feasibility testing
            feasible_found = False
            for _ in range(10):
                obs = np.random.randint(0, 2, size=self.num_items)
                best_obs = np.random.randint(0, 2, size=self.num_items)
                try:
                    removes, adds = llm_func(obs, self.profits, self.weights, self.capacities, 
                                             np.zeros((0, self.num_objectives)), best_obs)
                except TypeError:
                    removes, adds = llm_func(obs, self.profits, self.weights, self.capacities, 
                                             np.zeros((0, self.num_objectives)))
                except Exception as e:
                    continue
                if self._is_feasible(obs, removes, adds):
                    feasible_found = True
                    break
            
            if not feasible_found:
                print("[DEBUG] LLM move did not pass feasibility tests.")
                return None
            
            def llm_move_wrapper(observation, profits, weights, capacities, archive_objs, best_observation=None):
                try:
                    if archive_objs is None:
                        archive_objs = np.zeros((0, profits.shape[0]))
                    try:
                        return llm_func(observation, profits, weights, capacities, archive_objs, best_observation)
                    except TypeError:
                        return llm_func(observation, profits, weights, capacities, archive_objs)
                except Exception:
                    return [0], [0]
            
            return llm_move_wrapper
        except Exception as e:
            print("[DEBUG] LLM func integration failed:", e)
            return None
    
    def _is_feasible(self, observation, removes, adds):
        """Check if a move is feasible"""
        if self.weights is None or self.capacities is None or removes is None or adds is None:
            return False
        items = observation.copy()
        for idx in removes:
            if idx < 0 or idx >= len(items) or items[idx] != 1:
                return False
            items[idx] = 0
        for idx in adds:
            if idx < 0 or idx >= len(items) or items[idx] != 0:
                return False
            items[idx] = 1
        total_weights = np.sum(self.weights * items, axis=1)
        feasible = np.all(total_weights <= self.capacities)
        return feasible
    
    def calculate_diversity_score(self, current_objs, archive_objs):
        """Calculate how diverse the current solution is compared to the archive"""
        if archive_objs is None or len(archive_objs) == 0:
            return 1.0
        
        # Calculate minimum distance to archive solutions
        distances = np.linalg.norm(archive_objs - current_objs, axis=1)
        min_distance = np.min(distances)
        
        # Normalize to [0, 1] range
        normalized_distance = min(min_distance / 1000.0, 1.0)
        
        return normalized_distance
    
    def mutate(self, registry, mutation_rate=0.25):
        self.architecture.mutate(registry, mutation_rate=mutation_rate)
    
    def clone(self):
        new_agent = ModularMetaAgent(
            self.num_items, self.num_objectives, self.profits, self.weights, self.capacities
        )
        new_agent.architecture = self.architecture.clone()
        if self.best_observation is not None:
            new_agent.best_observation = self.best_observation.copy()
        new_agent.llm_code_history = list(self.llm_code_history)
        new_agent.llm_funcs = dict(self.llm_funcs)
        new_agent.llm_stats = {k: v.copy() for k, v in self.llm_stats.items()}
        new_agent.last_fitness = list(self.last_fitness)
        new_agent.move_history = list(self.move_history)
        new_agent.outcome_history = list(self.outcome_history)
        return new_agent
````

````python
%%cython
import numpy as np
import random
import requests
import json

class ModularAgentPopulationManager:
    def __init__(self, num_agents, num_items, num_objectives, profits, weights, capacities, llm_endpoint=None):
        self.num_agents = num_agents
        self.num_items = num_items
        self.num_objectives = num_objectives
        self.registry = MOVE_REGISTRY
        self.llm_endpoint = llm_endpoint
        self.agents = [
            ModularMetaAgent(
                num_items, num_objectives, profits, weights, capacities,
                move_types=random.sample(list(MOVE_REGISTRY.keys()), k=random.randint(2, len(MOVE_REGISTRY)))
            )
            for _ in range(num_agents)
        ]
        self.agent_stats = [{"fitness_sum": 0.0, "moves": 0} for _ in range(self.num_agents)]
        self.generation_count = 0
        self.archive_history = []
        self.elite_archive = []  # Track elite solutions across generations
        
    def select_agent_for_solution(self, solution_idx):
        return self.agents[solution_idx % self.num_agents]
    
    def report_agent_performance(self, agent_idx, solution_fitness):
        self.agent_stats[agent_idx]["fitness_sum"] += solution_fitness
        self.agent_stats[agent_idx]["moves"] += 1
    
    def gses_agent_level(self):
        """Enhanced GSES cycle with MHRE concepts"""
        avg_fitness = [s["fitness_sum"]/s["moves"] if s["moves"] > 0 else -1e9 
                       for s in self.agent_stats]
        sorted_idx = np.argsort(-np.array(avg_fitness))
        survivors = [self.agents[i] for i in sorted_idx[:max(2, self.num_agents//2)]]
        
        # Collect performance data for cooperative evolution
        performance_data = []
        for agent in self.agents:
            if hasattr(agent, 'move_history') and hasattr(agent, 'outcome_history'):
                performance_data.append({
                    'move_sequence': agent.move_history,
                    'outcomes': agent.outcome_history
                })
        
        # Apply cooperative evolution
        if performance_data:
            survivors = self.cooperative_evolution(survivors, performance_data)
        
        # Share best LLM moves among survivors
        best_llm_moves = {}
        for agent in survivors:
            for move_name, func in getattr(agent, "llm_funcs", {}).items():
                if move_name in getattr(agent, "llm_stats", {}):
                    stats = agent.llm_stats[move_name]
                    success_rate = stats["successes"] / max(1, stats["calls"])
                    if success_rate > 0.5:
                        best_llm_moves[move_name] = func
        
        for agent in survivors:
            for move_name, func in best_llm_moves.items():
                if move_name not in getattr(agent, "llm_funcs", {}):
                    agent.llm_funcs[move_name] = func
                    agent.llm_stats[move_name] = {"calls": 0, "successes": 0}
                    agent.architecture.move_types.append(move_name)
                    agent.architecture.move_probs = np.append(agent.architecture.move_probs, 0.1)
                    agent.architecture.move_probs /= np.sum(agent.architecture.move_probs)
        
        # Try to generate new LLM moves
        for agent in survivors:
            if self.llm_endpoint and random.random() < 0.7:
                code = self.generate_llm_move_code(agent)
                if code:
                    agent.add_llm_move(code)
                agent.prune_llm_moves()
        
        # Crossover and mutation
        new_agents = []
        while len(new_agents) + len(survivors) < self.num_agents:
            parent1, parent2 = random.sample(survivors, 2)
            child = parent1.clone()
            
            # Enhanced crossover
            moves1 = set(parent1.architecture.move_types)
            moves2 = set(parent2.architecture.move_types)
            moves_child = list(moves1 | moves2)
            
            if len(moves_child) > 2:
                moves_child = random.sample(moves_child, k=random.randint(2, len(moves_child)))
            
            child.architecture.move_types = moves_child
            child.architecture.move_counts = {k: 1 for k in moves_child}
            child.architecture.move_successes = {k: 1 for k in moves_child}
            child.architecture.move_deltas = {k: [] for k in moves_child}
            child.architecture.move_probs = np.ones(len(moves_child)) / len(moves_child)
            
            child.mutate(self.registry, mutation_rate=0.25)
            new_agents.append(child)
        
        self.agents = survivors + new_agents
        self.generation_count += 1
        
        # Apply architecture upgrade periodically
        if self.generation_count % 5 == 0:
            self.architecture_upgrade(self.agents)
        
        self.agent_stats = [{"fitness_sum": 0.0, "moves": 0} for _ in range(self.num_agents)]
    
    def cooperative_evolution(self, agents, performance_data):
        """Enhanced cooperative evolution with interaction analysis"""
        interaction_matrix = self.analyze_move_interactions(performance_data)
        suggestions = self.generate_cooperative_suggestions(interaction_matrix)
        
        for agent in agents:
            if random.random() < 0.3:
                self.apply_cooperative_suggestions(agent, suggestions)
        
        return agents
    
    def analyze_move_interactions(self, performance_data):
        """Analyze move interactions to identify cooperative patterns"""
        interaction_matrix = np.zeros((len(self.registry), len(self.registry)))
        
        for data in performance_data:
            move_sequence = data['move_sequence']
            outcomes = data['outcomes']
            
            for i in range(len(move_sequence) - 1):
                move1 = move_sequence[i]
                move2 = move_sequence[i + 1]
                
                if move1 in self.registry and move2 in self.registry:
                    idx1 = list(self.registry.keys()).index(move1)
                    idx2 = list(self.registry.keys()).index(move2)
                    
                    if outcomes[i] and outcomes[i + 1]:
                        interaction_matrix[idx1, idx2] += 1
        
        # Normalize
        row_sums = interaction_matrix.sum(axis=1)
        for i in range(len(row_sums)):
            if row_sums[i] > 0:
                interaction_matrix[i] /= row_sums[i]
        
        return interaction_matrix
    
    def generate_cooperative_suggestions(self, interaction_matrix):
        """Generate suggestions based on interaction analysis"""
        suggestions = []
        
        # Identify strong positive interactions
        strong_interactions = np.where(interaction_matrix > 0.7)
        for i, j in zip(strong_interactions[0], strong_interactions[1]):
            if i != j:
                move1 = list(self.registry.keys())[i]
                move2 = list(self.registry.keys())[j]
                suggestions.append({
                    'type': 'strengthen_interaction',
                    'move1': move1,
                    'move2': move2,
                    'reason': f"Strong positive interaction between {move1} and {move2}"
                })
        
        # Identify weak interactions
        weak_interactions = np.where((interaction_matrix > 0) & (interaction_matrix < 0.3))
        for i in weak_interactions[0]:
            move = list(self.registry.keys())[i]
            suggestions.append({
                'type': 'improve_move',
                'move': move,
                'reason': f"Weak interactions for {move}, consider improving"
            })
        
        return suggestions
    
    def apply_cooperative_suggestions(self, agent, suggestions):
        """Apply cooperative suggestions to agent"""
        for suggestion in suggestions:
            if suggestion['type'] == 'strengthen_interaction':
                move1 = suggestion['move1']
                move2 = suggestion['move2']
                
                if move1 in agent.architecture.move_types and move2 in agent.architecture.move_types:
                    idx1 = agent.architecture.move_types.index(move1)
                    idx2 = agent.architecture.move_types.index(move2)
                    
                    # Strengthen the interaction by increasing probabilities
                    agent.architecture.move_probs[idx1] *= 1.2
                    agent.architecture.move_probs[idx2] *= 1.2
                    agent.architecture.move_probs /= np.sum(agent.architecture.move_probs)
            
            elif suggestion['type'] == 'improve_move':
                move = suggestion['move']
                
                if move.startswith('llm_move_') and move in getattr(agent, "llm_funcs", {}):
                    current_code = None
                    for code in getattr(agent, "llm_code_history", []):
                        if f"def {move}" in code:
                            current_code = code
                            break
                    
                    if current_code:
                        improved_code = self.improve_llm_move_code(current_code, suggestion['reason'])
                        if improved_code:
                            agent.llm_funcs.pop(move, None)
                            agent.llm_stats.pop(move, None)
                            
                            if move in agent.architecture.move_types:
                                idx = agent.architecture.move_types.index(move)
                                agent.architecture.move_types.pop(idx)
                                agent.architecture.move_probs = np.delete(agent.architecture.move_probs, idx)
                                agent.architecture.move_probs /= np.sum(agent.architecture.move_probs)
                            
                            agent.add_llm_move(improved_code)
    
    def architecture_upgrade(self, agents):
        """Upgrade agent architectures based on elite patterns"""
        elite_agents = self.get_elite_agents(top_percent=0.2)
        common_patterns = self.analyze_elite_patterns(elite_agents)
        
        for agent in agents:
            if agent not in elite_agents and random.random() < 0.5:
                self.upgrade_agent_architecture(agent, common_patterns)
        
        return agents
    
    def get_elite_agents(self, top_percent=0.2):
        """Get elite agents based on performance"""
        avg_fitness = [s["fitness_sum"]/s["moves"] if s["moves"] > 0 else -1e9 
                       for s in self.agent_stats]
        elite_indices = np.argsort(-np.array(avg_fitness))[:int(len(self.agents) * top_percent)]
        return [self.agents[i] for i in elite_indices]
    
    def analyze_elite_patterns(self, elite_agents):
        """Analyze patterns in elite agents"""
        patterns = {
            'move_type_frequency': {},
            'move_sequence_patterns': [],
            'parameter_settings': {}
        }
        
        for agent in elite_agents:
            for move_type in agent.architecture.move_types:
                patterns['move_type_frequency'][move_type] = patterns['move_type_frequency'].get(move_type, 0) + 1
        
        for agent in elite_agents:
            patterns['move_sequence_patterns'].append(agent.architecture.move_types)
        
        for agent in elite_agents:
            for move_type, prob in zip(agent.architecture.move_types, agent.architecture.move_probs):
                if move_type not in patterns['parameter_settings']:
                    patterns['parameter_settings'][move_type] = []
                patterns['parameter_settings'][move_type].append(prob)
        
        return patterns
    
    def upgrade_agent_architecture(self, agent, patterns):
        """Upgrade agent architecture based on elite patterns"""
        for move_type in agent.architecture.move_types:
            if move_type in patterns['move_type_frequency']:
                elite_freq = patterns['move_type_frequency'][move_type] / len(patterns['move_type_frequency'])
                current_idx = agent.architecture.move_types.index(move_type)
                agent.architecture.move_probs[current_idx] *= (1 + elite_freq)
        
        agent.architecture.move_probs /= np.sum(agent.architecture.move_probs)
        
        # Add missing elite moves
        elite_moves = set(patterns['move_type_frequency'].keys())
        current_moves = set(agent.architecture.move_types)
        missing_moves = elite_moves - current_moves
        
        if missing_moves and random.random() < 0.7:
            move_to_add = random.choice(list(missing_moves))
            agent.architecture.move_types.append(move_to_add)
            agent.architecture.move_probs = np.append(agent.architecture.move_probs, 0.1)
            agent.architecture.move_probs /= np.sum(agent.architecture.move_probs)
    
    def generate_llm_move_code(self, agent):
        """Generate LLM move code with enhanced context"""
        agent_context = self.get_agent_performance_context(agent)
        archive_context = self.get_archive_context()
        
        prompt = f"""
You are an expert in multi-objective optimization and metaheuristics. Your task is to create a novel move function for a multi-objective knapsack problem.

CONTEXT:
- Number of items: {self.num_items}
- Number of objectives: {self.num_objectives}
- Current agent performance: {agent_context}
- Archive diversity: {archive_context}

REQUIREMENTS:
1. Function signature: def llm_move(observation, profits, weights, capacities, archive_objs, best_observation=None):
2. The function should return: (remove_indices: list[int], add_indices: list[int])
3. Focus on creating moves that improve diversity in the objective space
4. Consider using advanced techniques such as:
   - Information theory (entropy, mutual information)
   - Clustering analysis
   - Correlation between objectives
   - Novel selection strategies
5. Ensure the move is feasible (doesn't exceed knapsack capacities)
6. Try to propose multi-item moves (up to 3 items) when beneficial

PERFORMANCE GUIDANCE:
- The current agent has strengths in: {agent_context.get('strengths', 'unknown')}
- The agent struggles with: {agent_context.get('weaknesses', 'unknown')}
- Focus on addressing the weaknesses while maintaining strengths

EXAMPLE STRATEGIES:
1. Entropy-based selection: Calculate entropy of items and select based on information gain
2. Diversity-focused: Maximize minimum distance to archive solutions
3. Correlation-aware: Consider correlation between objectives when selecting items
4. Cluster-based: Group items by similarity and select from different clusters

Please provide a complete, well-commented Python function that implements a novel move strategy.
"""
        
        try:
            response = requests.post(
                self.llm_endpoint,
                json={"model": "llama3", "prompt": prompt, "stream": False}
            )
            text = response.json()["response"]
            
            if "def llm_move" in text:
                start_idx = text.find("def llm_move")
                next_def = text.find("\ndef ", start_idx + 1)
                if next_def == -1:
                    function_code = text[start_idx:]
                else:
                    function_code = text[start_idx:next_def]
                return function_code
            else:
                return None
        except Exception as e:
            print(f"[DEBUG] LLM error: {e}")
            return None
    
    def improve_llm_move_code(self, current_code, reason):
        """Improve an existing LLM move function"""
        prompt = f"""
Improve the following optimization function based on the performance feedback:

Function code:
```python
{current_code}
````

````python

### Cell 6: Enhanced Cython Integration with MHRE Framework
```python
%%cython
from libc.stdlib cimport malloc, free, srand, rand
from libc.string cimport memset
from libc.math cimport exp
import numpy as np
cimport numpy as np
import time

# Define structures
cdef struct ind:
    int nombr_nonpris
    int nombr
    int rank
    float fitnessbest
    float fitness
    int explored
    double *f
    double *capa
    double *v
    int *d
    int *Items

cdef struct pop:
    int size
    int maxsize
    ind **ind_array

# Global variables
cdef int NBITEMS = 250
cdef int ni = 250
cdef int L = 5
cdef double LARGE = 10e50
cdef float smallValue = 0.0000001
cdef double kappa = 0.05
cdef int alpha = 10
cdef int paretoIni = 28000
cdef int nf = 2
cdef double *capacities = NULL
cdef int **weights = NULL
cdef int **profits = NULL
cdef double *vector_weight = NULL
cdef double max_bound = 0.0
cdef double **OBJ_Weights = NULL
cdef int nombreLIGNE = 0
cdef int nextLn = 0
cdef int inv = 0
cdef int OBJ_Weights_lines = 0

# Python interface variables
cdef public object py_agent_manager = None
cdef public object move_stats = None

def set_agent_manager(agent_manager):
    global py_agent_manager
    py_agent_manager = agent_manager

def reset_move_stats():
    global move_stats
    move_stats = {
        "total_agent_calls": 0, 
        "move_type_counts": {}, 
        "total_time_agent": 0.0, 
        "total_time_loop": 0.0,
        "cache_hits": 0,
        "cache_misses": 0,
        "generation_count": 0,
        "llm_moves_generated": 0,
        "cooperative_evolutions": 0
    }

def print_move_stats():
    print("\n=== MHRE PERFORMANCE STATS ===")
    print("Total agent calls:", move_stats["total_agent_calls"])
    print("Move type counts:", move_stats["move_type_counts"])
    print("Total agent time (sec):", move_stats["total_time_agent"])
    print("Total loop time (sec):", move_stats["total_time_loop"])
    print("Cache hits:", move_stats["cache_hits"])
    print("Cache misses:", move_stats["cache_misses"])
    print("Generations:", move_stats["generation_count"])
    print("LLM moves generated:", move_stats["llm_moves_generated"])
    print("Cooperative evolutions:", move_stats["cooperative_evolutions"])
    
    if move_stats["total_agent_calls"] > 0:
        print("Avg agent call time (ms):", 1000.0 * move_stats["total_time_agent"] / move_stats["total_agent_calls"])

def seed(int x):
    srand(x)

cdef int irand(int range_val):
    return rand() % range_val

cdef void *chk_malloc(size_t size):
    cdef void *return_value = malloc(size)
    if return_value == NULL:
        raise MemoryError("Out of memory.")
    memset(return_value, 0, size)
    return return_value

cdef pop *create_pop(int maxsize, int nf):
    cdef int i
    cdef pop *pp = <pop *>chk_malloc(sizeof(pop))
    pp.size = 0
    pp.maxsize = maxsize
    pp.ind_array = <ind **>chk_malloc(maxsize * sizeof(void*))
    for i in range(maxsize):
        pp.ind_array[i] = NULL
    return pp

cdef ind *create_ind(int nf):
    cdef int i
    cdef ind *p_ind = <ind *>chk_malloc(sizeof(ind))
    p_ind.nombr_nonpris = 0
    p_ind.nombr = 0
    p_ind.rank = 0
    p_ind.fitnessbest = -1.0
    p_ind.fitness = -1.0
    p_ind.explored = 0
    p_ind.f = <double *>chk_malloc(nf * sizeof(double))
    p_ind.capa = <double *>chk_malloc(nf * sizeof(double))
    p_ind.v = <double *>chk_malloc(nf * sizeof(double))
    p_ind.d = <int *>chk_malloc(ni * sizeof(int))
    p_ind.Items = <int *>chk_malloc(ni * sizeof(int))
    for i in range(ni):
        p_ind.Items[i] = 0
        p_ind.d[i] = 0
    for i in range(nf):
        p_ind.f[i] = 0.0
        p_ind.capa[i] = 0.0
        p_ind.v[i] = 0.0
    return p_ind

cdef ind *ind_copy(ind *i):
    cdef ind *p_ind = create_ind(nf)
    cdef int k
    p_ind.nombr_nonpris = i.nombr_nonpris
    p_ind.nombr = i.nombr
    p_ind.rank = i.rank
    p_ind.fitnessbest = i.fitnessbest
    p_ind.fitness = i.fitness
    p_ind.explored = i.explored
    for k in range(nf):
        p_ind.f[k] = i.f[k]
        p_ind.v[k] = i.v[k]
        p_ind.capa[k] = i.capa[k]
    for k in range(ni):
        p_ind.d[k] = i.d[k]
        p_ind.Items[k] = i.Items[k]
    return p_ind

cdef void free_ind(ind *p_ind):
    if p_ind != NULL:
        free(p_ind.d)
        free(p_ind.f)
        free(p_ind.capa)
        free(p_ind.v)
        free(p_ind.Items)
        free(p_ind)

cdef void complete_free_pop(pop *pp):
    cdef int i
    if pp != NULL:
        if pp.ind_array != NULL:
            for i in range(pp.size):
                if pp.ind_array[i] != NULL:
                    free_ind(pp.ind_array[i])
                    pp.ind_array[i] = NULL
            free(pp.ind_array)
        free(pp)

cdef void cleanup_globals():
    global capacities, weights, profits, vector_weight, OBJ_Weights, OBJ_Weights_lines, nf, ni
    if capacities != NULL:
        free(capacities)
        capacities = NULL
    if weights != NULL:
        for i in range(nf):
            if weights[i] != NULL:
                free(weights[i])
        free(weights)
        weights = NULL
    if profits != NULL:
        for i in range(nf):
            if profits[i] != NULL:
                free(profits[i])
        free(profits)
        profits = NULL
    if vector_weight != NULL:
        free(vector_weight)
        vector_weight = NULL
    if OBJ_Weights != NULL:
        for i in range(nf):
            if OBJ_Weights[i] != NULL:
                free(OBJ_Weights[i])
        free(OBJ_Weights)
        OBJ_Weights = NULL
    OBJ_Weights_lines = 0
    nf = 0
    ni = 0

cdef int non_dominated(ind *p_ind_a, ind *p_ind_b):
    cdef int i
    cdef int a_is_good = -1
    cdef int equal = 1
    for i in range(nf):
        if p_ind_a.f[i] > p_ind_b.f[i]:
            a_is_good = 1
        if p_ind_a.f[i] != p_ind_b.f[i]:
            equal = 0
    if equal:
        return 0
    return a_is_good

cdef double calcAddEpsIndicator(ind *p_ind_a, ind *p_ind_b):
    global max_bound
    cdef int i
    cdef double eps
    cdef double temp_eps
    if max_bound == 0.0:
        max_bound = 1e-8
    eps = (p_ind_a.v[0]/max_bound)-(p_ind_b.v[0]/max_bound)
    for i in range(1, nf):
        temp_eps = (p_ind_a.v[i]/max_bound)-(p_ind_b.v[i]/max_bound)
        if temp_eps > eps:
            eps = temp_eps
    return eps

cdef void init_fitness(ind *x):
    x.fitness = 0.0

cdef void update_fitness(ind *x, double I):
    x.fitness -= exp(-I / kappa)

cdef double update_fitness_return(double f, double I):
    return f - exp(-I / kappa)

cdef int delete_fitness(ind *x, double I):
    x.fitness += exp(-I / kappa)
    return 0

cdef void compute_ind_fitness(ind *x, pop *SP):
    cdef int j
    init_fitness(x)
    for j in range(SP.size):
        if SP.ind_array[j] != x:
            update_fitness(x, calcAddEpsIndicator(SP.ind_array[j], x))

cdef void compute_all_fitness(pop *SP):
    cdef int i
    for i in range(SP.size):
        compute_ind_fitness(SP.ind_array[i])

cdef void loadMOKP(char *filename):
    global nf, ni, capacities, weights, profits
    cdef int i, f
    with open(filename.decode(), "r") as source:
        _nf, _ni = [int(x) for x in source.readline().split()]
        nf = _nf
        ni = _ni
        capacities = <double *>chk_malloc(nf * sizeof(double))
        weights = <int **>chk_malloc(nf * sizeof(void*))
        profits = <int **>chk_malloc(nf * sizeof(void*))
        for f in range(nf):
            capacities[f] = float(source.readline().strip())
            weights[f] = <int *>chk_malloc(ni * sizeof(int))
            profits[f] = <int *>chk_malloc(ni * sizeof(int))
            for i in range(ni):
                source.readline()
                weights[f][i] = int(source.readline().strip())
                profits[f][i] = int(source.readline().strip())

cdef void read_weights_file(char *filename):
    global OBJ_Weights, nombreLIGNE, nf, OBJ_Weights_lines
    cdef int i, j, nlines
    with open(filename.decode(), "r") as f:
        lines = [line for line in f if line.strip()]
    nlines = len(lines)
    OBJ_Weights = <double **>chk_malloc(nf * sizeof(void*))
    for i in range(nf):
        OBJ_Weights[i] = <double *>chk_malloc(nlines * sizeof(double))
    for i, line in enumerate(lines):
        vals = line.strip().split()
        for j in range(nf):
            OBJ_Weights[j][i] = float(vals[j])
    nombreLIGNE = nlines - 1
    OBJ_Weights_lines = nlines

cdef void dynamic_weight_allpop():
    global vector_weight, OBJ_Weights, nombreLIGNE, nf, nextLn
    cdef int i
    if vector_weight == NULL:
        vector_weight = <double *>chk_malloc(nf * sizeof(double))
    for i in range(nf):
        vector_weight[i] = OBJ_Weights[i][nextLn]
    if nextLn == nombreLIGNE:
        nextLn = 0
    else:
        nextLn += 1

cdef void choose_weight():
    dynamic_weight_allpop()

cdef void random_init_ind(ind *x):
    cdef int j, r, tmp
    for j in range(ni):
        x.d[j] = j
    for j in range(ni):
        r = irand(ni)
        tmp = x.d[r]
        x.d[r] = x.d[j]
        x.d[j] = tmp

cdef void evaluate(ind *x):
    cdef int j, l, k, faisable
    x.nombr = 0
    x.nombr_nonpris = 0
    for j in range(nf):
        x.capa[j] = 0.0
        x.f[j] = 0.0
    for j in range(ni):
        l = 0
        faisable = 1
        while l < nf and faisable == 1:
            if x.capa[l] + weights[l][x.d[j]] > capacities[l]:
                faisable = 0
            l += 1
        if faisable == 1:
            for k in range(nf):
                x.capa[k] += weights[k][x.d[j]]
                x.f[k] += profits[k][x.d[j]]
            x.Items[x.d[j]] = 1
            x.nombr += 1
        else:
            x.Items[x.d[j]] = 0
            x.nombr_nonpris += 1

cdef void P_init_pop(pop *SP, pop *Sarchive, int alpha):
    cdef int i, x, tmp, t
    t = max(alpha, Sarchive.size)
    cdef int* shuffle = <int *>chk_malloc(t * sizeof(int))
    for i in range(t):
        shuffle[i] = i
    for i in range(t):
        x = irand(alpha)
        tmp = shuffle[i]
        shuffle[i] = shuffle[x]
        shuffle[x] = tmp
    SP.size = alpha
    if Sarchive.size > alpha:
        for i in range(alpha):
            SP.ind_array[i] = ind_copy(Sarchive.ind_array[shuffle[i]])
    else:
        for i in range(alpha):
            if shuffle[i] < Sarchive.size:
                SP.ind_array[i] = ind_copy(Sarchive.ind_array[shuffle[i]])
            else:
                SP.ind_array[i] = create_ind(nf)
                random_init_ind(SP.ind_array[i])
                evaluate(SP.ind_array[i])
    free(shuffle)

cdef int extractPtoArchive(pop *P, pop *archive):
    cdef int i, j, dom, t, convergence_rate
    t = archive.size + P.size
    archiveAndP = create_pop(t, nf)
    convergence_rate = 0
    for i in range(archive.size):
        archiveAndP.ind_array[i] = archive.ind_array[i]
    for i in range(P.size):
        archiveAndP.ind_array[i + archive.size] = ind_copy(P.ind_array[i])
    archiveAndP.size = t
    archive.size = 0
    for i in range(t):
        for j in range(t):
            if i != j:
                dom = non_dominated(archiveAndP.ind_array[i], archiveAndP.ind_array[j])
                if dom == -1 or (dom == 0 and i > j):
                    break
        else:
            archive.ind_array[archive.size] = ind_copy(archiveAndP.ind_array[i])
            archive.size += 1
            if i >= t - P.size:
                convergence_rate += 1
    complete_free_pop(archiveAndP)
    return convergence_rate

cdef double calcMaxbound(pop *SP, int size):
    global max_bound
    cdef int i, j
    SP.size = size
    cdef double max_b = SP.ind_array[0].v[0]
    for i in range(SP.size):
        for j in range(nf):
            if max_b < SP.ind_array[i].v[j]:
                max_b = SP.ind_array[i].v[j]
    if max_b == 0.0:
        max_b = 1e-8
    max_bound = max_b
    return max_b

cdef void calcul_weight(pop *SP, int size):
    cdef int i, j
    for i in range(SP.size):
        for j in range(nf):
            SP.ind_array[i].v[j] = SP.ind_array[i].f[j] * vector_weight[j]

cdef int compute_fitness_and_select(pop *SP, ind *x, int size):
    cdef int i, worst
    cdef double worst_fit, fit_tmp
    SP.size = size
    x.fitness = 0
    compute_ind_fitness(x, SP)
    worst_fit = x.fitness
    worst = -1
    for i in range(SP.size):
        fit_tmp = update_fitness_return(SP.ind_array[i].fitness, calcAddEpsIndicator(x, SP.ind_array[i]))
        if fit_tmp > worst_fit:
            worst = i
            worst_fit = fit_tmp
    fit_tmp = x.fitness
    if worst == -1:
        return -1
    else:
        for i in range(SP.size):
            delete_fitness(SP.ind_array[i], calcAddEpsIndicator(SP.ind_array[worst], SP.ind_array[i]))
            update_fitness(SP.ind_array[i], calcAddEpsIndicator(x, SP.ind_array[i]))
        delete_fitness(x, calcAddEpsIndicator(SP.ind_array[worst], x))
        free_ind(SP.ind_array[worst])
        SP.ind_array[worst] = ind_copy(x)
        if fit_tmp - worst_fit > smallValue:
            return worst
        else:
            return -1

cdef np.ndarray get_archive_objs(pop *archive):
    cdef int i, j
    arr = np.zeros((archive.size, nf), dtype=np.float64)
    for i in range(archive.size):
        for j in range(nf):
            arr[i, j] = archive.ind_array[i].f[j]
    return arr

cdef np.ndarray get_items(ind *x):
    arr = np.zeros(ni, dtype=np.int32)
    cdef int i
    for i in range(ni):
        arr[i] = x.Items[i]
    return arr

# Built-in Cython move functions
cdef tuple cy_mutation_move(np.ndarray[np.int32_t, ndim=1] items, int ni):
    cdef int n_selected = 0, n_unselected = 0, i
    for i in range(ni):
        if items[i] == 1: n_selected += 1
        else: n_unselected += 1
    if n_selected == 0 or n_unselected == 0:
        return (0, 0)
    sel = [i for i in range(ni) if items[i] == 1]
    unsel = [i for i in range(ni) if items[i] == 0]
    remove_idx = sel[irand(n_selected)]
    add_idx = unsel[irand(n_unselected)]
    return remove_idx, add_idx

cdef tuple cy_local_search_move(np.ndarray[np.int32_t, ndim=1] items, double[:, :] profits, int ni, int nf):
    cdef int i
    sel = [i for i in range(ni) if items[i] == 1]
    unsel = [i for i in range(ni) if items[i] == 0]
    if not sel or not unsel:
        return (0, 0)
    profs_selected = [sum([profits[f, idx] for f in range(nf)]) for idx in sel]
    profs_unselected = [sum([profits[f, idx] for f in range(nf)]) for idx in unsel]
    remove_idx = sel[np.argmin(profs_selected)]
    add_idx = unsel[np.argmax(profs_unselected)]
    return remove_idx, add_idx

cdef tuple cy_global_search_move(np.ndarray[np.int32_t, ndim=1] items, double[:, :] profits, int ni, int nf):
    cdef int i
    sel = [i for i in range(ni) if items[i] == 1]
    unsel = [i for i in range(ni) if items[i] == 0]
    if not sel or not unsel:
        return (0, 0)
    profs_selected = np.array([sum([profits[f, idx] for f in range(nf)]) for idx in sel], dtype=np.float64) + 1e-9
    profs_unselected = np.array([sum([profits[f, idx] for f in range(nf)]) for idx in unsel], dtype=np.float64) + 1e-9
    remove_idx = sel[np.random.choice(len(sel), p=profs_selected/np.sum(profs_selected))]
    add_idx = unsel[np.random.choice(len(unsel), p=profs_unselected/np.sum(profs_unselected))]
    return remove_idx, add_idx

# Example LLM move (replace with your best LLM move logic)
cdef tuple cy_llm_move_0(np.ndarray[np.int32_t, ndim=1] items, double[:, :] profits, int ni, int nf):
    """
    Example Cythonized LLM move: removes 2 least profitable selected, adds 2 most profitable unselected
    """
    sel = [i for i in range(ni) if items[i] == 1]
    unsel = [i for i in range(ni) if items[i] == 0]
    if len(sel) < 2 or len(unsel) < 2:
        return (0, 0)
    profs_selected = [sum([profits[f, idx] for f in range(nf)]) for idx in sel]
    profs_unselected = [sum([profits[f, idx] for f in range(nf)]) for idx in unsel]
    remove_idx = [sel[i] for i in np.argsort(profs_selected)[:2]]
    add_idx = [unsel[i] for i in np.argsort(profs_unselected)[-2:]]
    return remove_idx, add_idx

def get_profits_np():
    global profits, nf, ni
    arr = np.zeros((nf, ni), dtype=np.float64)
    for f in range(nf):
        for i in range(ni):
            arr[f, i] = profits[f][i]
    return arr

cdef void Indicator_local_search1(pop *SP, pop *Sarchive, int size):
    cdef ind *x
    cdef ind *y
    cdef int i, j, r, t, k, l, v, sol, mino, mp, maxp, consistant, pos, stop, convergence, ii, tmp_pris, tmp_nonpris, taille, feasible, tv, IM
    cdef int* remplace = <int *>chk_malloc(L * sizeof(int))
    cdef np.ndarray[np.int32_t, ndim=1] items_mv = np.empty(ni, dtype=np.int32)
    
    SP.size = size
    extractPtoArchive(SP, Sarchive)
    profits_np = get_profits_np()
    
    while True:
        convergence = 0
        archive_objs = get_archive_objs(Sarchive) if Sarchive.size > 0 else None
        
        # Update archive history for agent manager
        if py_agent_manager is not None and archive_objs is not None:
            py_agent_manager.archive_history.append(archive_objs.copy())
            if len(py_agent_manager.archive_history) > 100:
                py_agent_manager.archive_history = py_agent_manager.archive_history[-100:]
        
        loop_start_time = time.time()
        
        for i in range(SP.size):
            if not SP.ind_array[i].explored:
                x = ind_copy(SP.ind_array[i])
                j = 0
                while j < x.nombr:
                    for l in range(L):
                        remplace[l] = 0
                    for ii in range(ni):
                        items_mv[ii] = x.Items[ii]
                    
                    # Enhanced move selection with MHRE concepts
                    move_types = ["mutation", "local_search", "global_search", "follow", "diversity", "llm_move_0"]
                    
                    # Use agent manager for move selection
                    if py_agent_manager is not None:
                        agent_idx = i % py_agent_manager.num_agents
                        agent = py_agent_manager.agents[agent_idx]
                        
                        state = {
                            "Items": [x.Items[ii] for ii in range(ni)],
                            "capa": [x.capa[ii] for ii in range(nf)],
                            "f": [x.f[ii] for ii in range(nf)],
                        }
                        
                        context = {}
                        if archive_objs is not None:
                            context["archive_objs"] = archive_objs
                        
                        # Get best observation from archive
                        if Sarchive.size > 0:
                            best_idx = 0
                            best_fitness = -LARGE
                            for idx in range(Sarchive.size):
                                fitness = sum([Sarchive.ind_array[idx].f[f] for f in range(nf)])
                                if fitness > best_fitness:
                                    best_fitness = fitness
                                    best_idx = idx
                            context["best_observation"] = get_items(Sarchive.ind_array[best_idx])
                        
                        try:
                            obs = agent.observe(state)
                            removes, adds = agent.act(obs, context)
                            agent_move_type = agent.last_move_type
                        except Exception as e:
                            print("Agent error:", e)
                            removes, adds = [], []
                            agent_move_type = "error"
                    else:
                        # Fallback to random selection
                        mt_idx = irand(len(move_types))
                        move_type = move_types[mt_idx]
                        agent_move_type = move_type
                        
                        # Built-in moves
                        if move_type == "mutation":
                            remove_idx, add_idx = cy_mutation_move(items_mv, ni)
                            removes, adds = [remove_idx], [add_idx]
                        elif move_type == "local_search":
                            remove_idx, add_idx = cy_local_search_move(items_mv, profits_np, ni, nf)
                            removes, adds = [remove_idx], [add_idx]
                        elif move_type == "global_search":
                            remove_idx, add_idx = cy_global_search_move(items_mv, profits_np, ni, nf)
                            removes, adds = [remove_idx], [add_idx]
                        elif move_type == "llm_move_0":
                            removes, adds = cy_llm_move_0(items_mv, profits_np, ni, nf)
                        else:
                            removes, adds = [0], [0]
                    
                    agent_start_time = time.time()
                    
                    # Fallback logic
                    if not removes or not all(isinstance(idx, int) or isinstance(idx, np.integer) for idx in removes):
                        removes = []
                        while True:
                            mino = irand(ni)
                            if x.Items[mino] == 1:
                                removes = [mino]
                                break
                    if not adds or not all(isinstance(idx, int) or isinstance(idx, np.integer) for idx in adds):
                        adds = []
                        while True:
                            maxp = irand(ni)
                            if x.Items[maxp] == 0 and (not removes or maxp != removes[0]):
                                adds = [maxp]
                                break
                    
                    # Apply all removes
                    for remove_idx in removes:
                        if remove_idx < 0 or remove_idx >= ni or x.Items[remove_idx] != 1:
                            continue
                        x.Items[remove_idx] = 0
                        x.nombr -= 1
                        x.nombr_nonpris += 1
                        for r in range(nf):
                            x.capa[r] -= weights[r][remove_idx]
                            x.f[r] -= profits[r][remove_idx]
                    
                    IM = 0
                    taille = 0
                    while IM < L:
                        # Apply all adds
                        for add_idx in adds:
                            if add_idx < 0 or add_idx >= ni or x.Items[add_idx] != 0 or (removes and add_idx in removes):
                                continue
                            consistant = 1
                            r = 0
                            while r < nf and consistant == 1:
                                if x.capa[r] + weights[r][add_idx] > capacities[r]:
                                    consistant = 0
                                r += 1
                            if consistant == 1:
                                feasible = 1
                                r = 0
                                while r < taille and feasible:
                                    if add_idx == remplace[r]:
                                        feasible = 0
                                    r += 1
                                if feasible == 1:
                                    remplace[taille] = add_idx
                                    taille += 1
                                    x.Items[add_idx] = 1
                                    x.nombr_nonpris -= 1
                                    x.nombr += 1
                                    for r in range(nf):
                                        x.capa[r] += weights[r][add_idx]
                                        x.f[r] += profits[r][add_idx]
                        IM += 1
                    
                    for tv in range(nf):
                        x.v[tv] = x.f[tv] * vector_weight[tv]
                    max_bound = calcMaxbound(SP, SP.size)
                    sol = compute_fitness_and_select(SP, x, SP.size)
                    agent_success = (sol != -1)
                    
                    agent_end_time = time.time()
                    if move_stats is not None:
                        move_stats["total_agent_calls"] += 1
                        move_stats["total_time_agent"] += (agent_end_time - agent_start_time)
                        mt = agent_move_type if agent_move_type is not None else "unknown"
                        if mt not in move_stats["move_type_counts"]:
                            move_stats["move_type_counts"][mt] = 0
                        move_stats["move_type_counts"][mt] += 1
                    
                    if py_agent_manager is not None:
                        try:
                            agent.report_move_result(agent_move_type, agent_success)
                            fitness = np.sum([x.f[j] for j in range(nf)])
                            py_agent_manager.report_agent_performance(agent_idx, fitness)
                        except Exception:
                            pass
                    
                    if sol != -1:
                        j = x.nombr + 1
                        if sol > i and i + 1 < SP.size:
                            y = SP.ind_array[i + 1]
                            SP.ind_array[i + 1] = SP.ind_array[sol]
                            SP.ind_array[sol] = y
                            i += 1
                        break
                    elif sol == -1:
                        # Undo all removes
                        for remove_idx in removes:
                            x.Items[remove_idx] = 1
                            x.nombr_nonpris -= 1
                            x.nombr += 1
                            for r in range(nf):
                                x.capa[r] += weights[r][remove_idx]
                                x.f[r] += profits[r][remove_idx]
                        if taille >= 1:
                            for r in range(taille):
                                x.Items[remplace[r]] = 0
                                x.nombr -= 1
                                x.nombr_nonpris += 1
                                for t in range(nf):
                                    x.capa[t] -= weights[t][remplace[r]]
                                    x.f[t] -= profits[t][remplace[r]]
                                    x.v[t] = x.f[t] * vector_weight[t]
                    j += 1
                tmp_pris = x.nombr
                tmp_nonpris = x.nombr_nonpris
                free_ind(x)
                if j == tmp_pris:
                    SP.ind_array[i].explored = 1
        
        loop_end_time = time.time()
        if move_stats is not None:
            move_stats["total_time_loop"] += (loop_end_time - loop_start_time)
        
        convergence = extractPtoArchive(SP, Sarchive)
        if not convergence:
            break
    
    free(remplace)

def run_moacp(instance_file, weights_file, nbitems, num_objectives, output_file):
    global nf, ni, NBITEMS, alpha, paretoIni, L, nombreLIGNE, nextLn, inv, vector_weight
    global capacities, weights, profits, OBJ_Weights
    
    alpha = 10
    paretoIni = 28000
    NBL = 25
    NRUNS = 3
    
    reset_move_stats()
    
    for run in range(1, NRUNS+1):
        NBITEMS = nbitems
        ni = nbitems
        nf = num_objectives
        print(f"RUN {run}/{NRUNS} -- {instance_file.decode()} nbitems={ni} nf={nf} => {output_file}")
        nombreLIGNE = 0
        nextLn = 0
        inv = 0
        seed(run)
        loadMOKP(instance_file)
        read_weights_file(weights_file)
        vector_weight = <double *>chk_malloc(nf * sizeof(double))
        P = create_pop(paretoIni, nf)
        
        it = 0
        while it < NBL:
            solutions = create_pop(alpha, nf)
            archive = create_pop(paretoIni, nf)
            choose_weight()
            P_init_pop(solutions, P, alpha)
            extractPtoArchive(solutions, P)
            calcul_weight(solutions, alpha)
            calcMaxbound(solutions, alpha)
            compute_all_fitness(solutions)
            
            # Enhanced local search with MHRE
            Indicator_local_search1(solutions, archive, alpha)
            extractPtoArchive(archive, P)
            
            # Agent population GSES every 5 iterations
            if py_agent_manager is not None and hasattr(py_agent_manager, "gses_agent_level") and it % 5 == 0:
                py_agent_manager.gses_agent_level()
                if move_stats is not None:
                    move_stats["generation_count"] += 1
                    move_stats["cooperative_evolutions"] += 1
            
            # Track LLM moves generated
            if py_agent_manager is not None:
                for agent in py_agent_manager.agents:
                    if move_stats is not None:
                        move_stats["llm_moves_generated"] += len(agent.llm_funcs)
            
            it += 1
            complete_free_pop(solutions)
            complete_free_pop(archive)
        
        # Write results
        with open(output_file, "a") as fpareto:
            fpareto.write("\n")
            for i in range(P.size):
                for j in range(nf):
                    fpareto.write(f"{P.ind_array[i].f[j]:.6f} ")
                fpareto.write("\n")
        
        complete_free_pop(P)
        cleanup_globals()
        print_move_stats()
````

````python

### Cell 6: Cython Integration and Core Optimization Functions
```python
%%cython
from libc.stdlib cimport malloc, free, srand, rand
from libc.string cimport memset
from libc.math cimport exp
import numpy as np
cimport numpy as np
import time

# Define structures
cdef struct ind:
    int nombr_nonpris
    int nombr
    int rank
    float fitnessbest
    float fitness
    int explored
    double *f
    double *capa
    double *v
    int *d
    int *Items

cdef struct pop:
    int size
    int maxsize
    ind **ind_array

# Global variables
cdef int NBITEMS = 250
cdef int ni = 250
cdef int L = 5
cdef double LARGE = 10e50
cdef float smallValue = 0.0000001
cdef double kappa = 0.05
cdef int alpha = 10
cdef int paretoIni = 28000
cdef int nf = 2
cdef double *capacities = NULL
cdef int **weights = NULL
cdef int **profits = NULL
cdef double *vector_weight = NULL
cdef double max_bound = 0.0
cdef double **OBJ_Weights = NULL
cdef int nombreLIGNE = 0
cdef int nextLn = 0
cdef int inv = 0
cdef int OBJ_Weights_lines = 0

# Python interface variables
cdef public object py_agent_manager = None
cdef public object move_stats = None

def set_agent_manager(agent_manager):
    global py_agent_manager
    py_agent_manager = agent_manager

def reset_move_stats():
    global move_stats
    move_stats = {
        "total_agent_calls": 0, 
        "move_type_counts": {}, 
        "total_time_agent": 0.0, 
        "total_time_loop": 0.0,
        "cache_hits": 0,
        "cache_misses": 0
    }

def print_move_stats():
    print("\n=== AGENT MOVE PROFILE ===")
    print("Total agent calls:", move_stats["total_agent_calls"])
    print("Move type counts:", move_stats["move_type_counts"])
    print("Total agent time (sec):", move_stats["total_time_agent"])
    print("Total loop time (sec):", move_stats["total_time_loop"])
    print("Cache hits:", move_stats["cache_hits"])
    print("Cache misses:", move_stats["cache_misses"])
    if move_stats["total_agent_calls"] > 0:
        print("Avg agent call time (ms):", 1000.0 * move_stats["total_time_agent"] / move_stats["total_agent_calls"])

def seed(int x):
    srand(x)

cdef int irand(int range_val):
    return rand() % range_val

cdef void *chk_malloc(size_t size):
    cdef void *return_value = malloc(size)
    if return_value == NULL:
        raise MemoryError("Out of memory.")
    memset(return_value, 0, size)
    return return_value

cdef pop *create_pop(int maxsize, int nf):
    cdef int i
    cdef pop *pp = <pop *>chk_malloc(sizeof(pop))
    pp.size = 0
    pp.maxsize = maxsize
    pp.ind_array = <ind **>chk_malloc(maxsize * sizeof(void*))
    for i in range(maxsize):
        pp.ind_array[i] = NULL
    return pp

cdef ind *create_ind(int nf):
    cdef int i
    cdef ind *p_ind = <ind *>chk_malloc(sizeof(ind))
    p_ind.nombr_nonpris = 0
    p_ind.nombr = 0
    p_ind.rank = 0
    p_ind.fitnessbest = -1.0
    p_ind.fitness = -1.0
    p_ind.explored = 0
    p_ind.f = <double *>chk_malloc(nf * sizeof(double))
    p_ind.capa = <double *>chk_malloc(nf * sizeof(double))
    p_ind.v = <double *>chk_malloc(nf * sizeof(double))
    p_ind.d = <int *>chk_malloc(ni * sizeof(int))
    p_ind.Items = <int *>chk_malloc(ni * sizeof(int))
    for i in range(ni):
        p_ind.Items[i] = 0
        p_ind.d[i] = 0
    for i in range(nf):
        p_ind.f[i] = 0.0
        p_ind.capa[i] = 0.0
        p_ind.v[i] = 0.0
    return p_ind

cdef ind *ind_copy(ind *i):
    cdef ind *p_ind = create_ind(nf)
    cdef int k
    p_ind.nombr_nonpris = i.nombr_nonpris
    p_ind.nombr = i.nombr
    p_ind.rank = i.rank
    p_ind.fitnessbest = i.fitnessbest
    p_ind.fitness = i.fitness
    p_ind.explored = i.explored
    for k in range(nf):
        p_ind.f[k] = i.f[k]
        p_ind.v[k] = i.v[k]
        p_ind.capa[k] = i.capa[k]
    for k in range(ni):
        p_ind.d[k] = i.d[k]
        p_ind.Items[k] = i.Items[k]
    return p_ind

cdef void free_ind(ind *p_ind):
    if p_ind != NULL:
        free(p_ind.d)
        free(p_ind.f)
        free(p_ind.capa)
        free(p_ind.v)
        free(p_ind.Items)
        free(p_ind)

cdef void complete_free_pop(pop *pp):
    cdef int i
    if pp != NULL:
        if pp.ind_array != NULL:
            for i in range(pp.size):
                if pp.ind_array[i] != NULL:
                    free_ind(pp.ind_array[i])
                    pp.ind_array[i] = NULL
            free(pp.ind_array)
        free(pp)

cdef void cleanup_globals():
    global capacities, weights, profits, vector_weight, OBJ_Weights, OBJ_Weights_lines, nf, ni
    if capacities != NULL:
        free(capacities)
        capacities = NULL
    if weights != NULL:
        for i in range(nf):
            if weights[i] != NULL:
                free(weights[i])
        free(weights)
        weights = NULL
    if profits != NULL:
        for i in range(nf):
            if profits[i] != NULL:
                free(profits[i])
        free(profits)
        profits = NULL
    if vector_weight != NULL:
        free(vector_weight)
        vector_weight = NULL
    if OBJ_Weights != NULL:
        for i in range(nf):
            if OBJ_Weights[i] != NULL:
                free(OBJ_Weights[i])
        free(OBJ_Weights)
        OBJ_Weights = NULL
    OBJ_Weights_lines = 0
    nf = 0
    ni = 0

cdef int non_dominated(ind *p_ind_a, ind *p_ind_b):
    cdef int i
    cdef int a_is_good = -1
    cdef int equal = 1
    for i in range(nf):
        if p_ind_a.f[i] > p_ind_b.f[i]:
            a_is_good = 1
        if p_ind_a.f[i] != p_ind_b.f[i]:
            equal = 0
    if equal:
        return 0
    return a_is_good

cdef double calcAddEpsIndicator(ind *p_ind_a, ind *p_ind_b):
    global max_bound
    cdef int i
    cdef double eps
    cdef double temp_eps
    if max_bound == 0.0:
        max_bound = 1e-8
    eps = (p_ind_a.v[0]/max_bound)-(p_ind_b.v[0]/max_bound)
    for i in range(1, nf):
        temp_eps = (p_ind_a.v[i]/max_bound)-(p_ind_b.v[i]/max_bound)
        if temp_eps > eps:
            eps = temp_eps
    return eps

cdef void init_fitness(ind *x):
    x.fitness = 0.0

cdef void update_fitness(ind *x, double I):
    x.fitness -= exp(-I / kappa)

cdef double update_fitness_return(double f, double I):
    return f - exp(-I / kappa)

cdef int delete_fitness(ind *x, double I):
    x.fitness += exp(-I / kappa)
    return 0

cdef void compute_ind_fitness(ind *x, pop *SP):
    cdef int j
    init_fitness(x)
    for j in range(SP.size):
        if SP.ind_array[j] != x:
            update_fitness(x, calcAddEpsIndicator(SP.ind_array[j], x))

cdef void compute_all_fitness(pop *SP):
    cdef int i
    for i in range(SP.size):
        compute_ind_fitness(SP.ind_array[i])

cdef void loadMOKP(char *filename):
    global nf, ni, capacities, weights, profits
    cdef int i, f
    with open(filename.decode(), "r") as source:
        _nf, _ni = [int(x) for x in source.readline().split()]
        nf = _nf
        ni = _ni
        capacities = <double *>chk_malloc(nf * sizeof(double))
        weights = <int **>chk_malloc(nf * sizeof(void*))
        profits = <int **>chk_malloc(nf * sizeof(void*))
        for f in range(nf):
            capacities[f] = float(source.readline().strip())
            weights[f] = <int *>chk_malloc(ni * sizeof(int))
            profits[f] = <int *>chk_malloc(ni * sizeof(int))
            for i in range(ni):
                source.readline()
                weights[f][i] = int(source.readline().strip())
                profits[f][i] = int(source.readline().strip())

cdef void read_weights_file(char *filename):
    global OBJ_Weights, nombreLIGNE, nf, OBJ_Weights_lines
    cdef int i, j, nlines
    with open(filename.decode(), "r") as f:
        lines = [line for line in f if line.strip()]
    nlines = len(lines)
    OBJ_Weights = <double **>chk_malloc(nf * sizeof(void*))
    for i in range(nf):
        OBJ_Weights[i] = <double *>chk_malloc(nlines * sizeof(double))
    for i, line in enumerate(lines):
        vals = line.strip().split()
        for j in range(nf):
            OBJ_Weights[j][i] = float(vals[j])
    nombreLIGNE = nlines - 1
    OBJ_Weights_lines = nlines

cdef void dynamic_weight_allpop():
    global vector_weight, OBJ_Weights, nombreLIGNE, nf, nextLn
    cdef int i
    if vector_weight == NULL:
        vector_weight = <double *>chk_malloc(nf * sizeof(double))
    for i in range(nf):
        vector_weight[i] = OBJ_Weights[i][nextLn]
    if nextLn == nombreLIGNE:
        nextLn = 0
    else:
        nextLn += 1

cdef void choose_weight():
    dynamic_weight_allpop()

cdef void random_init_ind(ind *x):
    cdef int j, r, tmp
    for j in range(ni):
        x.d[j] = j
    for j in range(ni):
        r = irand(ni)
        tmp = x.d[r]
        x.d[r] = x.d[j]
        x.d[j] = tmp

cdef void evaluate(ind *x):
    cdef int j, l, k, faisable
    x.nombr = 0
    x.nombr_nonpris = 0
    for j in range(nf):
        x.capa[j] = 0.0
        x.f[j] = 0.0
    for j in range(ni):
        l = 0
        faisable = 1
        while l < nf and faisable == 1:
            if x.capa[l] + weights[l][x.d[j]] > capacities[l]:
                faisable = 0
            l += 1
        if faisable == 1:
            for k in range(nf):
                x.capa[k] += weights[k][x.d[j]]
                x.f[k] += profits[k][x.d[j]]
            x.Items[x.d[j]] = 1
            x.nombr += 1
        else:
            x.Items[x.d[j]] = 0
            x.nombr_nonpris += 1

cdef void P_init_pop(pop *SP, pop *Sarchive, int alpha):
    cdef int i, x, tmp, t
    t = max(alpha, Sarchive.size)
    cdef int* shuffle = <int *>chk_malloc(t * sizeof(int))
    for i in range(t):
        shuffle[i] = i
    for i in range(t):
        x = irand(alpha)
        tmp = shuffle[i]
        shuffle[i] = shuffle[x]
        shuffle[x] = tmp
    SP.size = alpha
    if Sarchive.size > alpha:
        for i in range(alpha):
            SP.ind_array[i] = ind_copy(Sarchive.ind_array[shuffle[i]])
    else:
        for i in range(alpha):
            if shuffle[i] < Sarchive.size:
                SP.ind_array[i] = ind_copy(Sarchive.ind_array[shuffle[i]])
            else:
                SP.ind_array[i] = create_ind(nf)
                random_init_ind(SP.ind_array[i])
                evaluate(SP.ind_array[i])
    free(shuffle)

cdef int extractPtoArchive(pop *P, pop *archive):
    cdef int i, j, dom, t, convergence_rate
    t = archive.size + P.size
    archiveAndP = create_pop(t, nf)
    convergence_rate = 0
    for i in range(archive.size):
        archiveAndP.ind_array[i] = archive.ind_array[i]
    for i in range(P.size):
        archiveAndP.ind_array[i + archive.size] = ind_copy(P.ind_array[i])
    archiveAndP.size = t
    archive.size = 0
    for i in range(t):
        for j in range(t):
            if i != j:
                dom = non_dominated(archiveAndP.ind_array[i], archiveAndP.ind_array[j])
                if dom == -1 or (dom == 0 and i > j):
                    break
        else:
            archive.ind_array[archive.size] = ind_copy(archiveAndP.ind_array[i])
            archive.size += 1
            if i >= t - P.size:
                convergence_rate += 1
    complete_free_pop(archiveAndP)
    return convergence_rate

cdef double calcMaxbound(pop *SP, int size):
    global max_bound
    cdef int i, j
    SP.size = size
    cdef double max_b = SP.ind_array[0].v[0]
    for i in range(SP.size):
        for j in range(nf):
            if max_b < SP.ind_array[i].v[j]:
                max_b = SP.ind_array[i].v[j]
    if max_b == 0.0:
        max_b = 1e-8
    max_bound = max_b
    return max_b

cdef void calcul_weight(pop *SP, int size):
    cdef int i, j
    for i in range(SP.size):
        for j in range(nf):
            SP.ind_array[i].v[j] = SP.ind_array[i].f[j] * vector_weight[j]

cdef int compute_fitness_and_select(pop *SP, ind *x, int size):
    cdef int i, worst
    cdef double worst_fit, fit_tmp
    SP.size = size
    x.fitness = 0
    compute_ind_fitness(x, SP)
    worst_fit = x.fitness
    worst = -1
    for i in range(SP.size):
        fit_tmp = update_fitness_return(SP.ind_array[i].fitness, calcAddEpsIndicator(x, SP.ind_array[i]))
        if fit_tmp > worst_fit:
            worst = i
            worst_fit = fit_tmp
    fit_tmp = x.fitness
    if worst == -1:
        return -1
    else:
        for i in range(SP.size):
            delete_fitness(SP.ind_array[i], calcAddEpsIndicator(SP.ind_array[worst], SP.ind_array[i]))
            update_fitness(SP.ind_array[i], calcAddEpsIndicator(x, SP.ind_array[i]))
        delete_fitness(x, calcAddEpsIndicator(SP.ind_array[worst], x))
        free_ind(SP.ind_array[worst])
        SP.ind_array[worst] = ind_copy(x)
        if fit_tmp - worst_fit > smallValue:
            return worst
        else:
            return -1

cdef np.ndarray get_archive_objs(pop *archive):
    cdef int i, j
    arr = np.zeros((archive.size, nf), dtype=np.float64)
    for i in range(archive.size):
        for j in range(nf):
            arr[i, j] = archive.ind_array[i].f[j]
    return arr

cdef np.ndarray get_items(ind *x):
    arr = np.zeros(ni, dtype=np.int32)
    cdef int i
    for i in range(ni):
        arr[i] = x.Items[i]
    return arr

# Built-in Cython move functions
cdef tuple cy_mutation_move(np.ndarray[np.int32_t, ndim=1] items, int ni):
    cdef int n_selected = 0, n_unselected = 0, i
    for i in range(ni):
        if items[i] == 1: n_selected += 1
        else: n_unselected += 1
    if n_selected == 0 or n_unselected == 0:
        return (0, 0)
    sel = [i for i in range(ni) if items[i] == 1]
    unsel = [i for i in range(ni) if items[i] == 0]
    remove_idx = sel[irand(n_selected)]
    add_idx = unsel[irand(n_unselected)]
    return remove_idx, add_idx

cdef tuple cy_local_search_move(np.ndarray[np.int32_t, ndim=1] items, double[:, :] profits, int ni, int nf):
    cdef int i
    sel = [i for i in range(ni) if items[i] == 1]
    unsel = [i for i in range(ni) if items[i] == 0]
    if not sel or not unsel:
        return (0, 0)
    profs_selected = [sum([profits[f, idx] for f in range(nf)]) for idx in sel]
    profs_unselected = [sum([profits[f, idx] for f in range(nf)]) for idx in unsel]
    remove_idx = sel[np.argmin(profs_selected)]
    add_idx = unsel[np.argmax(profs_unselected)]
    return remove_idx, add_idx

cdef tuple cy_global_search_move(np.ndarray[np.int32_t, ndim=1] items, double[:, :] profits, int ni, int nf):
    cdef int i
    sel = [i for i in range(ni) if items[i] == 1]
    unsel = [i for i in range(ni) if items[i] == 0]
    if not sel or not unsel:
        return (0, 0)
    profs_selected = np.array([sum([profits[f, idx] for f in range(nf)]) for idx in sel], dtype=np.float64) + 1e-9
    profs_unselected = np.array([sum([profits[f, idx] for f in range(nf)]) for idx in unsel], dtype=np.float64) + 1e-9
    remove_idx = sel[np.random.choice(len(sel), p=profs_selected/np.sum(profs_selected))]
    add_idx = unsel[np.random.choice(len(unsel), p=profs_unselected/np.sum(profs_unselected))]
    return remove_idx, add_idx

# Example LLM move (replace with your best LLM move logic)
cdef tuple cy_llm_move_0(np.ndarray[np.int32_t, ndim=1] items, double[:, :] profits, int ni, int nf):
    """
    Example Cythonized LLM move: removes 2 least profitable selected, adds 2 most profitable unselected
    """
    sel = [i for i in range(ni) if items[i] == 1]
    unsel = [i for i in range(ni) if items[i] == 0]
    if len(sel) < 2 or len(unsel) < 2:
        return (0, 0)
    profs_selected = [sum([profits[f, idx] for f in range(nf)]) for idx in sel]
    profs_unselected = [sum([profits[f, idx] for f in range(nf)]) for idx in unsel]
    remove_idx = [sel[i] for i in np.argsort(profs_selected)[:2]]
    add_idx = [unsel[i] for i in np.argsort(profs_unselected)[-2:]]
    return remove_idx, add_idx

def get_profits_np():
    global profits, nf, ni
    arr = np.zeros((nf, ni), dtype=np.float64)
    for f in range(nf):
        for i in range(ni):
            arr[f, i] = profits[f][i]
    return arr

cdef void Indicator_local_search1(pop *SP, pop *Sarchive, int size):
    cdef ind *x
    cdef ind *y
    cdef int i, j, r, t, k, l, v, sol, mino, mp, maxp, consistant, pos, stop, convergence, ii, tmp_pris, tmp_nonpris, taille, feasible, tv, IM
    cdef int* remplace = <int *>chk_malloc(L * sizeof(int))
    cdef np.ndarray[np.int32_t, ndim=1] items_mv = np.empty(ni, dtype=np.int32)
    
    SP.size = size
    extractPtoArchive(SP, Sarchive)
    profits_np = get_profits_np()
    
    while True:
        convergence = 0
        archive_objs = get_archive_objs(Sarchive) if Sarchive.size > 0 else None
        loop_start_time = time.time()
        
        for i in range(SP.size):
            if not SP.ind_array[i].explored:
                x = ind_copy(SP.ind_array[i])
                j = 0
                while j < x.nombr:
                    for l in range(L):
                        remplace[l] = 0
                    for ii in range(ni):
                        items_mv[ii] = x.Items[ii]
                    
                    # Select move type
                    move_types = ["mutation", "local_search", "global_search", "follow", "diversity", "llm_move_0"]
                    mt_idx = irand(len(move_types))
                    move_type = move_types[mt_idx]
                    agent_move_type = move_type
                    agent_success = False
                    agent_start_time = time.time()
        
                    removes = []
                    adds = []
                    # Built-in moves
                    if move_type == "mutation":
                        remove_idx, add_idx = cy_mutation_move(items_mv, ni)
                        removes, adds = [remove_idx], [add_idx]
                    elif move_type == "local_search":
                        remove_idx, add_idx = cy_local_search_move(items_mv, profits_np, ni, nf)
                        removes, adds = [remove_idx], [add_idx]
                    elif move_type == "global_search":
                        remove_idx, add_idx = cy_global_search_move(items_mv, profits_np, ni, nf)
                        removes, adds = [remove_idx], [add_idx]
                    # LLM move
                    elif move_type == "llm_move_0":
                        removes, adds = cy_llm_move_0(items_mv, profits_np, ni, nf)
                    # Agent moves
                    elif (move_type == "diversity" or move_type == "follow") and py_agent_manager is not None:
                        agent_idx = i % py_agent_manager.num_agents
                        agent = py_agent_manager.agents[agent_idx]
                        state = {
                            "Items": [x.Items[ii] for ii in range(ni)],
                            "capa": [x.capa[ii] for ii in range(nf)],
                            "f": [x.f[ii] for ii in range(nf)],
                        }
                        context = {}
                        if archive_objs is not None:
                            context["archive_objs"] = archive_objs
                        try:
                            obs = agent.observe(state)
                            removes, adds = agent.act(obs, context)
                            if hasattr(agent, "last_move_type"):
                                agent_move_type = agent.last_move_type
                        except Exception as e:
                            print("Agent error:", e)
                            removes, adds = [], []
                            agent_move_type = "error"
                    agent_end_time = time.time()
                    if move_stats is not None:
                        move_stats["total_agent_calls"] += 1
                        move_stats["total_time_agent"] += (agent_end_time - agent_start_time)
                        mt = agent_move_type if agent_move_type is not None else "unknown"
                        if mt not in move_stats["move_type_counts"]:
                            move_stats["move_type_counts"][mt] = 0
                        move_stats["move_type_counts"][mt] += 1
        
                    # Fallback logic
                    if not removes or not all(isinstance(idx, int) or isinstance(idx, np.integer) for idx in removes):
                        removes = []
                        while True:
                            mino = irand(ni)
                            if x.Items[mino] == 1:
                                removes = [mino]
                                break
                    if not adds or not all(isinstance(idx, int) or isinstance(idx, np.integer) for idx in adds):
                        adds = []
                        while True:
                            maxp = irand(ni)
                            if x.Items[maxp] == 0 and (not removes or maxp != removes[0]):
                                adds = [maxp]
                                break
        
                    # Apply all removes
                    for remove_idx in removes:
                        if remove_idx < 0 or remove_idx >= ni or x.Items[remove_idx] != 1:
                            continue
                        x.Items[remove_idx] = 0
                        x.nombr -= 1
                        x.nombr_nonpris += 1
                        for r in range(nf):
                            x.capa[r] -= weights[r][remove_idx]
                            x.f[r] -= profits[r][remove_idx]
        
                    IM = 0
                    taille = 0
                    while IM < L:
                        # Apply all adds
                        for add_idx in adds:
                            if add_idx < 0 or add_idx >= ni or x.Items[add_idx] != 0 or (removes and add_idx in removes):
                                continue
                            consistant = 1
                            r = 0
                            while r < nf and consistant == 1:
                                if x.capa[r] + weights[r][add_idx] > capacities[r]:
                                    consistant = 0
                                r += 1
                            if consistant == 1:
                                feasible = 1
                                r = 0
                                while r < taille and feasible:
                                    if add_idx == remplace[r]:
                                        feasible = 0
                                    r += 1
                                if feasible == 1:
                                    remplace[taille] = add_idx
                                    taille += 1
                                    x.Items[add_idx] = 1
                                    x.nombr_nonpris -= 1
                                    x.nombr += 1
                                    for r in range(nf):
                                        x.capa[r] += weights[r][add_idx]
                                        x.f[r] += profits[r][add_idx]
                        IM += 1
        
                    for tv in range(nf):
                        x.v[tv] = x.f[tv] * vector_weight[tv]
                    max_bound = calcMaxbound(SP, SP.size)
                    sol = compute_fitness_and_select(SP, x, SP.size)
                    agent_success = (sol != -1)
                    if py_agent_manager is not None and (move_type == "diversity" or move_type == "follow"):
                        try:
                            agent.report_move_result(agent_move_type, agent_success)
                        except Exception:
                            pass
                    if py_agent_manager is not None:
                        agent_idx = i % py_agent_manager.num_agents
                        fitness = np.sum([x.f[j] for j in range(nf)])
                        py_agent_manager.report_agent_performance(agent_idx, fitness)
                    if sol != -1:
                        j = x.nombr + 1
                        if sol > i and i + 1 < SP.size:
                            y = SP.ind_array[i + 1]
                            SP.ind_array[i + 1] = SP.ind_array[sol]
                            SP.ind_array[sol] = y
                            i += 1
                        break
                    elif sol == -1:
                        # Undo all removes
                        for remove_idx in removes:
                            x.Items[remove_idx] = 1
                            x.nombr_nonpris -= 1
                            x.nombr += 1
                            for r in range(nf):
                                x.capa[r] += weights[r][remove_idx]
                                x.f[r] += profits[r][remove_idx]
                        if taille >= 1:
                            for r in range(taille):
                                x.Items[remplace[r]] = 0
                                x.nombr -= 1
                                x.nombr_nonpris += 1
                                for t in range(nf):
                                    x.capa[t] -= weights[t][remplace[r]]
                                    x.f[t] -= profits[t][remplace[r]]
                                    x.v[t] = x.f[t] * vector_weight[t]
                    j += 1
                tmp_pris = x.nombr
                tmp_nonpris = x.nombr_nonpris
                free_ind(x)
                if j == tmp_pris:
                    SP.ind_array[i].explored = 1
        loop_end_time = time.time()
        if move_stats is not None:
            move_stats["total_time_loop"] += (loop_end_time - loop_start_time)
        convergence = extractPtoArchive(SP, Sarchive)
        if not convergence:
            break
    free(remplace)

def run_moacp(instance_file, weights_file, nbitems, num_objectives, output_file):
    global nf, ni, NBITEMS, alpha, paretoIni, L, nombreLIGNE, nextLn, inv, vector_weight
    global capacities, weights, profits, OBJ_Weights
    
    alpha = 10
    paretoIni = 28000
    NBL = 25
    NRUNS = 3
    
    reset_move_stats()
    
    for run in range(1, NRUNS+1):
        NBITEMS = nbitems
        ni = nbitems
        nf = num_objectives
        print(f"RUN {run}/{NRUNS} -- {instance_file.decode()} nbitems={ni} nf={nf} => {output_file}")
        nombreLIGNE = 0
        nextLn = 0
        inv = 0
        seed(run)
        loadMOKP(instance_file)
        read_weights_file(weights_file)
        vector_weight = <double *>chk_malloc(nf * sizeof(double))
        P = create_pop(paretoIni, nf)
        it = 0
        while it < NBL:
            solutions = create_pop(alpha, nf)
            archive = create_pop(paretoIni, nf)
            choose_weight()
            P_init_pop(solutions, P, alpha)
            extractPtoArchive(solutions, P)
            calcul_weight(solutions, alpha)
            calcMaxbound(solutions, alpha)
            compute_all_fitness(solutions)
            Indicator_local_search1(solutions, archive, alpha)
            extractPtoArchive(archive, P)
            
            # Agent population GSES every 10 iterations
            if py_agent_manager is not None and hasattr(py_agent_manager, "gses_agent_level") and it % 10 == 0:
                py_agent_manager.gses_agent_level()
            
            it += 1
            complete_free_pop(solutions)
            complete_free_pop(archive)
        
        # Write results
        with open(output_file, "a") as fpareto:
            fpareto.write("\n")
            for i in range(P.size):
                for j in range(nf):
                    fpareto.write(f"{P.ind_array[i].f[j]:.6f} ")
                fpareto.write("\n")
        
        complete_free_pop(P)
        cleanup_globals()
        print_move_stats()
````

````python
# --- Parameters and File Names ---
nbitems = 250
num_objectives = 2
instance_file = b"250.2.txt"
weights_file = b"Weights_2obj_FQ200.txt"
output_file = "kss5_improved.txt"
random_seed = 42
api_key = None  # Set your OpenAI API key here if you want to use LLM features

# Set random seed for reproducibility
random.seed(random_seed)
np.random.seed(random_seed)

# Initialize move_stats globally
move_stats = None

# 1. Load your instance as usual
print("Loading problem instance...")
profits = load_profits_from_file("250.2.txt", num_objectives, nbitems)
weights = load_weights_from_file("250.2.txt", num_objectives, nbitems)
capacities = load_capacities_from_file("250.2.txt", num_objectives, nbitems)
print(f"Loaded instance with {nbitems} items and {num_objectives} objectives")

# 2. Create LLM generator if API key is provided
llm_generator = None
if api_key:
    print("Initializing LLM generator...")
    llm_generator = LLMHeuristicGenerator(api_key)
else:
    print("No API key provided. Running without LLM features.")

# 3. Create the enhanced modular MHRE agent population
print("Creating agent population manager...")
agent_manager = EnhancedModularAgentPopulationManager(
    num_agents=10,
    num_items=nbitems,
    num_objectives=num_objectives,
    profits=profits,
    weights=weights,
    capacities=capacities,
    llm_generator=llm_generator
)

# 4. Register agent manager with cython
print("Registering agent manager with Cython...")
set_agent_manager(agent_manager)

# 5. Run the optimizer
print("Starting optimization...")
try:
    run_moacp(instance_file, weights_file, nbitems, num_objectives, output_file, api_key)
    print("Optimization completed successfully.")
except Exception as e:
    print(f"Optimization failed with error: {e}")
    import traceback
    traceback.print_exc()

# --- Visualization and Analysis ---
print("\nGenerating visualizations...")

import matplotlib.pyplot as plt

def load_pareto_front(filename):
    points = []
    try:
        with open(filename) as f:
            for line in f:
                line = line.strip()
                if line:
                    points.append([float(x) for x in line.split()])
    except FileNotFoundError:
        print(f"Warning: File {filename} not found")
        return np.array([])
    return np.array(points)

def is_dominated(point, others):
    """Returns True if point is dominated by any row in others"""
    if len(others) == 0:
        return False
    return np.any(np.all(others >= point, axis=1) & np.any(others > point, axis=1))

def get_pareto_front(points):
    if len(points) == 0:
        return np.array([])
    mask = np.ones(len(points), dtype=bool)
    for i, p in enumerate(points):
        others = np.delete(points, i, axis=0)
        if is_dominated(p, others):
            mask[i] = False
    return points[mask]

def calculate_hypervolume(points, reference_point=None):
    """Calculate hypervolume of the Pareto front"""
    if len(points) == 0:
        return 0.0
    
    # If no reference point provided, use the worst values
    if reference_point is None:
        reference_point = [max(p[i] for p in points) for i in range(len(points[0]))]
    
    # For 2 objectives, we can use a simple calculation
    if len(points[0]) == 2:
        # Sort by first objective
        sorted_points = sorted(points, key=lambda x: x[0])
        
        volume = 0.0
        prev_x = reference_point[0]
        for p in sorted_points:
            volume += (prev_x - p[0]) * (reference_point[1] - p[1])
            prev_x = p[0]
        
        return volume
    else:
        # For more objectives, use a library like pyhv
        try:
            import pyhv
            return pyhv.hypervolume(points, reference_point)
        except ImportError:
            print("pyhv library not available. Cannot calculate hypervolume for >2 objectives.")
            return 0.0

def calculate_igd(approximation, reference):
    """Calculate Inverted Generational Distance"""
    if len(approximation) == 0 or len(reference) == 0:
        return float('inf')
    
    # Calculate minimum distance from each reference point to any approximation point
    total_distance = 0.0
    for ref_point in reference:
        min_dist = min(
            np.sqrt(sum((a - b) ** 2 for a, b in zip(ref_point, approx_point))) 
            for approx_point in approximation
        )
        total_distance += min_dist
    
    return total_distance / len(reference)

def calculate_spread(points):
    """Calculate the spread of solutions in the Pareto front"""
    if len(points) < 2:
        return 0.0
    
    # Calculate the distance between consecutive solutions
    distances = []
    sorted_points = sorted(points)
    
    for i in range(len(sorted_points) - 1):
        dist = np.sqrt(sum((a - b) ** 2 for a, b in zip(sorted_points[i], sorted_points[i+1])))
        distances.append(dist)
    
    # Calculate average and standard deviation
    avg_distance = sum(distances) / len(distances)
    std_distance = np.sqrt(sum((d - avg_distance) ** 2 for d in distances) / len(distances))
    
    return std_distance / avg_distance if avg_distance > 0 else 0.0

# Load and merge all runs' points for each result file
print("Loading results...")
try:
    ref_points = load_pareto_front("2502_Resulats.txt")
    new_points = load_pareto_front("kss5_improved.txt")
    
    # Get the true Pareto front for each
    ref_pareto = get_pareto_front(ref_points)
    new_pareto = get_pareto_front(new_points)
    
    # Calculate performance metrics
    ref_hv = calculate_hypervolume(ref_pareto)
    new_hv = calculate_hypervolume(new_pareto)
    
    # Calculate IGD if reference is available
    if len(ref_pareto) > 0 and len(new_pareto) > 0:
        igd = calculate_igd(new_pareto, ref_pareto)
    else:
        igd = float('inf')
    
    # Calculate spread
    ref_spread = calculate_spread(ref_pareto)
    new_spread = calculate_spread(new_pareto)
    
    print(f"\nPerformance Comparison:")
    print(f"Reference Pareto Front - Size: {len(ref_pareto)}, Hypervolume: {ref_hv:.2f}, Spread: {ref_spread:.4f}")
    print(f"Improved Pareto Front - Size: {len(new_pareto)}, Hypervolume: {new_hv:.2f}, Spread: {new_spread:.4f}")
    print(f"IGD: {igd:.4f}")
    if ref_hv > 0:
        print(f"Hypervolume Improvement: {((new_hv - ref_hv) / ref_hv * 100):.2f}%")
    
    # Plot Pareto fronts
    plt.figure(figsize=(12, 10))
    
    # Main comparison plot
    plt.subplot(2, 2, 1)
    if len(ref_pareto) > 0:
        plt.scatter(ref_pareto[:,0], ref_pareto[:,1], label="Reference True Pareto", alpha=0.7, s=50)
    if len(new_pareto) > 0:
        plt.scatter(new_pareto[:,0], new_pareto[:,1], label="Improved True Pareto", alpha=0.7, s=50)
    plt.xlabel("Objective 1")
    plt.ylabel("Objective 2")
    plt.legend()
    plt.title("Pareto Front Comparison")
    plt.grid(True)
    
    # Hypervolume comparison
    plt.subplot(2, 2, 2)
    methods = ['Reference', 'Improved']
    volumes = [ref_hv, new_hv]
    bars = plt.bar(methods, volumes, color=['blue', 'orange'])
    plt.ylabel('Hypervolume')
    plt.title('Hypervolume Comparison')
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height:.2f}',
                 ha='center', va='bottom')
    
    # Spread comparison
    plt.subplot(2, 2, 3)
    spreads = [ref_spread, new_spread]
    bars = plt.bar(methods, spreads, color=['blue', 'orange'])
    plt.ylabel('Spread')
    plt.title('Solution Spread Comparison')
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height:.4f}',
                 ha='center', va='bottom')
    
    # Performance metrics summary
    plt.subplot(2, 2, 4)
    plt.axis('off')
    metrics_text = f"""
    Performance Metrics:
    
    Reference:
    - Solutions: {len(ref_pareto)}
    - Hypervolume: {ref_hv:.2f}
    - Spread: {ref_spread:.4f}
    
    Improved:
    - Solutions: {len(new_pareto)}
    - Hypervolume: {new_hv:.2f}
    - Spread: {new_spread:.4f}
    - IGD: {igd:.4f}
    
    HV Improvement: {((new_hv - ref_hv) / ref_hv * 100):.2f}%
    """
    plt.text(0.1, 0.9, metrics_text, transform=plt.gca().transAxes, 
             fontsize=10, verticalalignment='top', fontfamily='monospace')
    
    plt.tight_layout()
    plt.savefig("performance_analysis.png", dpi=300)
    plt.show()
    
    print("Performance analysis saved to 'performance_analysis.png'")
    
except Exception as e:
    print(f"Error during visualization: {e}")
    import traceback
    traceback.print_exc()

# Try to get move_stats from the Cython module
try:
    # Import the Cython module to access move_stats
    import sys
    # The Cython module name might be different, but we can try to access it
    # This is a workaround since we can't directly import the %%cython module
    # We'll try to access it through the global namespace
    if 'move_stats' in globals():
        current_move_stats = globals()['move_stats']
    else:
        current_move_stats = None
except:
    current_move_stats = None

# Plot move statistics if available
if current_move_stats is not None:
    print("\nGenerating move statistics visualization...")
    try:
        plt.figure(figsize=(15, 6))
        
        # Move type distribution
        plt.subplot(1, 3, 1)
        move_types = list(current_move_stats["move_type_counts"].keys())
        counts = list(current_move_stats["move_type_counts"].values())
        plt.bar(move_types, counts)
        plt.title("Move Type Distribution")
        plt.xlabel("Move Type")
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        
        # Performance over generations
        plt.subplot(1, 3, 2)
        if current_move_stats["generation_stats"]:
            generations = range(1, len(current_move_stats["generation_stats"]) + 1)
            calls = [stats["total_calls"] for stats in current_move_stats["generation_stats"]]
            plt.plot(generations, calls, 'o-')
            plt.title("Agent Calls Over Generations")
            plt.xlabel("Generation")
            plt.ylabel("Total Agent Calls")
        
        # Time efficiency
        plt.subplot(1, 3, 3)
        if current_move_stats["total_agent_calls"] > 0:
            avg_time = 1000.0 * current_move_stats["total_time_agent"] / current_move_stats["total_agent_calls"]
            total_time = current_move_stats["total_time_loop"]
            
            categories = ['Avg Agent Call (ms)', 'Total Loop Time (s)']
            values = [avg_time, total_time]
            bars = plt.bar(categories, values, color=['green', 'purple'])
            plt.title("Time Efficiency")
            plt.ylabel("Time")
            for bar in bars:
                height = bar.get_height()
                plt.text(bar.get_x() + bar.get_width()/2., height,
                         f'{height:.2f}',
                         ha='center', va='bottom')
        
        plt.tight_layout()
        plt.savefig("move_statistics.png", dpi=300)
        plt.show()
        
        print("Move statistics saved to 'move_statistics.png'")
        
    except Exception as e:
        print(f"Error generating move statistics: {e}")
else:
    print("No move statistics available")

# Print final summary
print("\n" + "="*50)
print("OPTIMIZATION SUMMARY")
print("="*50)
if current_move_stats is not None:
    print(f"Total agent calls: {current_move_stats['total_agent_calls']}")
    print(f"Total optimization time: {current_move_stats['total_time_loop']:.2f} seconds")
    if current_move_stats['total_agent_calls'] > 0:
        print(f"Average agent call time: {1000.0 * current_move_stats['total_time_agent'] / current_move_stats['total_agent_calls']:.2f} ms")
    print("\nMove type usage:")
    for move_type, count in current_move_stats["move_type_counts"].items():
        percentage = (count / current_move_stats["total_agent_calls"]) * 100
        print(f"  {move_type}: {count} calls ({percentage:.1f}%)")
else:
    print("No move statistics available (optimization may have failed)")

print("\nOptimization complete. Check the generated PNG files for visualizations.")
````
